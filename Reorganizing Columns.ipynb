### A notebook dedicated to reorganizing the columns in the training data, thus creating x_test_reorganized_columns.csv and x_train_reorganized_columns.csv

```python
import pandas as pd
import os
```

```python
data = pd.read_csv('data/trainingdata/extensive_x_train.csv')
```

```python
data = data.drop(columns = ['Date','Volume_1d_change'])
```

```python
data
```

```python
all_columns = data.columns
```

```python
real_time_columns = all_columns[:9]
```

```python
real_time_columns
```

```python
all_columns[:9]
```

```python
real_time_columns = reversed(real_time_columns)
```

A loop that creates a list of all the columns in the required order:

```python
new_index = []
for column in real_time_columns:
    for number in range(21):
        if number == 0:
            column_name = str(column)
        else:
            column_name = str(column) +'lag' + str(number)
        new_index.append(column_name)
```

```python
len(new_index)
```

```python
new_index
```

```python
reorganized_data = data.reindex(columns = new_index)
```

```python
reorganized_data.to_csv("data/trainingdata/x_train_reorganized_columns.csv")
```

```python
x_test_data = pd.read_csv('data/trainingdata/extensive_x_test.csv')
```

```python
x_test_data = x_test_data.drop(columns = ['Date','Volume_1d_change'])
```

```python
reorganized_x_test_data = x_test_data.reindex(columns=new_index)
```

```python
reorganized_x_test_data.to_csv("data/trainingdata/x_test_reorganized_columns.csv")
```
