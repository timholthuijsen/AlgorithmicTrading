# LSTM for Stock Price prediction
### A dedicated notebook for the creation of a Long Short-Term Memory (LSTM) Neural Network for stock price prediction and pattern recognition

```python
import tensorflow as tf
import os
import numpy as np
import pandas as pd
from importlib import reload
from sklearn.preprocessing import StandardScaler
from keras.layers import Dropout
```

### Data organization

```python
import ImportTrainingData
#loading my own training data
x_train = pd.read_csv('data/trainingdata/x_train_reorganized_columns.csv')
x_test = pd.read_csv('data/trainingdata/x_test_reorganized_columns.csv')
y_train = ImportTrainingData.y_train
y_test = ImportTrainingData.y_test
stock_data = ImportTrainingData.stock_data
```

```python
#convert them to numpy arrays
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

```python
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 9 #['ma14', 'rsi14', 'ma30', 'rsi30', 'ma50', 'rsi50', 'ma200', 'rsi200','Adj Close']
x_train = x_train.reshape((x_train.shape[0],21, n_features))
x_test = x_test.reshape((x_test.shape[0],21,n_features))
x_train.shape
```

```python
#normalize the training data:
scalers = {}
for i in range(x_train.shape[1]):
    scalers[i] = StandardScaler()
    x_train[:, i, :] = scalers[i].fit_transform(x_train[:, i, :]) 

for i in range(x_test.shape[1]):
    x_test[:, i, :] = scalers[i].transform(x_test[:, i, :]) 


y_scaler = StandardScaler().fit(y_train)
y_train = y_scaler.transform(y_train)
y_test = y_scaler.transform(y_test)
```

## Model creation
Now that everything all the data is in the correct shape, we can attempt to formulate a simple LSTM model to predict stock prices:

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
 

# define model
model = Sequential()
model.add(LSTM(50, activation='elu',return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(LSTM(50,activation='elu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(x_train, y_train, epochs=200, verbose=1)



yhat = model.predict(x_test, verbose=0)
#print(yhat)
```

And eveluate the model:

```python
mape = tf.keras.losses.MeanAbsolutePercentageError()
mean_percentage_error = mape(y_test,yhat).numpy()
print(f'mean percentage error: {mean_percentage_error}')

mae = tf.keras.losses.MeanAbsoluteError()
mean_absolute_error = mae(y_test,yhat).numpy()
print(f'mean absolute error: {mean_absolute_error}')
```

Now that we have defined an initial model and see that it can predict prices with some skill, we attempt to tune the hyperparameters and architecture of the model in order to create the optimal model

```python
def tune_model(neurons, activation, epochs=200):
    # define model
    model = Sequential()
    model.add(LSTM(neurons[0], activation=activation,return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
    #for neuron in neurons:
    for neuron in neurons:
        model.add(LSTM(neuron,activation=activation, return_sequences = True))
    model.add(LSTM(neurons[0],activation=activation))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mae')
    # fit model
    model.fit(x_train, y_train, epochs=epochs, verbose=0)

    yhat = model.predict(x_test, verbose=0)

    mape = tf.keras.losses.MeanAbsolutePercentageError()
    mean_percentage_error = mape(y_test,yhat).numpy()
    mae = tf.keras.losses.MeanAbsoluteError()
    mean_absolute_error = mae(y_test,yhat).numpy()
    return mean_absolute_error, mean_percentage_error
```

## Hyperparameter results
We have done some hyperparameter tuning. The most effective hyperparameters so far are:
neurons:[10, 20], activation:'elu'
This is likely due to 'elu' being a good activation. The 2 neurons are also strikingly low. We will experiment more with low neurons, and elu activation:

```python
possible_activations = ['elu']
neuron_combinations = [[1],[2,2],[2,3,4,5,6],[3,4,5,12],[3,4,4],[3,3,4,3,4],[40,3,4],[50,70,120]]

def hyperparameter_tuning(possible_activations, neuron_combinations):
    counter = 1
    results = []

    for activation in possible_activations:
        for neurons in neuron_combinations:
            print('at iteration', counter, 'out of', len(possible_activations)*len(neuron_combinations))
            counter+=1
            absolute, percentage = tune_model(neurons,activation,epochs=500)
            result = [str(absolute),str(percentage),'are the results for setup with neurons:', str(neurons), 'and activation:',activation]
            results.append(result)

    print(results)
    return results

#This is where the actual tuning happens:
#newest_results = hyperparameter_tuning(possible_activations, neuron_combinations)
#print(newest_results)
```
